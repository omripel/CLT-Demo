```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#U(0,10) 
reps = 1000000
for samples in ([1,2,5,10,20,30,50, 100]):
    sns.histplot([np.mean(np.random.uniform(0, 10,samples)) for j in range(reps)])
    plt.xlim(0, 10)
    plt.title("num of samples is %f" % samples)
    plt.show()
```

```python
#exp(1/2)
reps = 1000000

for samples in ([1,2,5,10,20,30,50, 100]):
    sns.histplot([np.mean(np.random.exponential(5, samples)) for j in range(reps)])
    plt.xlim(0, 10)
    plt.title("num of samples is %f" % samples)
    plt.show()
```
